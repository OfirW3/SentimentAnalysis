#KNN שיסווג את הטקסטים על פי רגשם

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

df.dropna(inplace=True)
tfidf = TfidfVectorizer(max_features=1000)  # Limit features for efficiency
df = df[df['sentiment'].isin(['positive', 'negative'])]  # Filter for positive and negative sentiments
X = tfidf.fit_transform(df['text']).toarray()  # Convert text to TF-IDF features so the KNN can understand

# Step 2: Encode categorical sentiment labels (target in y)
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['sentiment'])  # Encode sentiment labels

# Step 3: Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=16)
# Step 4: Find the best K - its 24

# Step 5: Train a KNN classifier
knn = KNeighborsClassifier(n_neighbors=int(24))
knn.fit(X_train, y_train)

# Step 6: Evaluate the model
accuracy = knn.score(X_test, y_test)
print(f"Accuracy: {accuracy:}")
````

````python
# prompt: Find the best K for the KNN model above

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from wordcloud import WordCloud, STOPWORDS
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the data (assuming 'df' is already loaded as in your previous code)
# ... (your existing code to load and preprocess the data)

# ... (Your existing code to create and train the model)
df.dropna(inplace=True)
tfidf = TfidfVectorizer(max_features=1000)  # Limit features for efficiency
df = df[df['sentiment'].isin(['positive', 'negative'])]  # Filter for positive and negative sentiments
X = tfidf.fit_transform(df['text']).toarray()  # Convert text to TF-IDF features so the KNN can understand

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['sentiment'])  # Encode sentiment labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=16)

# Find the best K
best_k = 1
best_accuracy = 0

for k in range(1, 50):  # Test values of K from 1 to 49
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)
  y_pred = knn.predict(X_test)
  accuracy = accuracy_score(y_test, y_pred)
  if accuracy > best_accuracy:
    best_accuracy = accuracy
    best_k = k

print(f"Best K: {best_k}")
print(f"Best Accuracy: {best_accuracy}")

#Train with best K
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train, y_train)
accuracy = knn.score(X_test, y_test)
print(f"Accuracy with best k: {accuracy}")
````
